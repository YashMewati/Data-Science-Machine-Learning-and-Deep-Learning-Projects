# Exploratory Data Analysis-PLOT

This Seaborn plot shows the Variation Between Confirmed Cases in Indian National & Foreign National 

```python
plt.figure(figsize=(18,10))
sns.displot(x='ConfirmedIndianNational',y='ConfirmedForeignNational',data=c19)
```

Here we study the Plot Between Confirmed and Cured !  

```python
sns.relplot(x='Confirmed',y='Cured',data=c19)
```

Some more EDA Between Confirmed and Cured Cases, varries From Different States !

```python
plt.figure(figsize=(18,10))
sns.relplot(x='Confirmed',y='Cured',hue='State/UnionTerritory',data=c19)
```

# Droping Columns - Sno, Date , Time 
Droping columns form our datasets to get more visual connections Between Confirmed,cured and Deaths

```python
c19drop=c19.drop(['Sno','Date','Time'],axis=1)
```

```python
c19drop.columns
```

This Barplot Shows The Deaths and Cured Ratio 

```python
plt.figure(figsize=(18,10))
sns.barplot(x='Deaths' ,y='Cured',data=c19drop)
```

```python
plt.figure(figsize=(18,10))
sns.histplot(x='Deaths',data=c19,palette='coolwarm',color='Red',bins=20)
```
